```python
from pathlib import Path
import pandas as pd
from astromodule.io import read_table, write_table
from astromodule.table import crossmatch, fast_crossmatch
from astromodule.datamodel import DESI
```

```python
TABLE_NAME = 'properties_interacting.dat'

df_int = read_table(Path('astromorphlib') / TABLE_NAME)
df_morpho = read_table('astromorphlib/gz_desi_deep_learning_catalog_friendly.parquet')
```

```python
df_morpho.columns
```

```python
cols = [
  'smooth-or-featured_smooth_fraction', 
  'smooth-or-featured_featured-or-disk_fraction',
  'disk-edge-on_yes_fraction',
  'disk-edge-on_no_fraction',
  'has-spiral-arms_no_fraction',
  'bar_strong_fraction',
  'bulge-size_dominant_fraction',
  'how-rounded_cigar-shaped_fraction',
  'edge-on-bulge_boxy_fraction',
  'spiral-winding_tight_fraction',
  'spiral-winding_loose_fraction',
  'spiral-arm-count_1_fraction',
  'merging_major-disturbance_fraction',
  'merging_merger_fraction',
]
# df_morpho = df_morpho.fillna(0)
for n in cols:
  sorted_df = df_morpho.sort_values(n, ascending=False, axis='index')
  sorted_df = sorted_df[['ra', 'dec', n]]
  sorted_df = sorted_df.iloc[:45]
  write_table(sorted_df, f'astromorphlib/questions/{n}.csv')
```

```python
columns = [
  'ra',
  'dec',
  'z',
  DESI.MERGING.MERGER.fr,
  DESI.MERGING.MAJOR_DISTURBANCE.fr,
  DESI.MERGING.MINOR_DISTURBANCE.fr,
  DESI.MERGING.NONE.fr,
  DESI.DISK_EDGE_ON.YES.fr,
]
C_max = 4
n_max = 2
F_max = 0.5
df_int['C_lim'] = -9.50 * df_int.A + 4.85
df_int['n_lim'] = -4.75 * df_int.A + 1.73
df_int['F_lim'] = -4.75 * df_int.A + 0.73
df_int = df_int[
  (df_int.C > df_int.C_lim) & (df_int.C < C_max) &
  # (df_int.N > df_int.n_lim) & (df_int.N < n_max) &
  (df_int['F(G_M20)'] > df_int.F_lim) & (df_int['F(G_M20)'] < F_max)
]
df_match = fast_crossmatch(df_int, df_morpho)
# df_match = df_match[columns]
write_table(df_match, 'astromorphlib/interacting+morph.csv')
df_match
```

```python
from astromodule.legacy import LegacyService
from astromodule.io import read_table
from astromodule.utils import iauname_path
from astromodule.datamodel import DESI


df = read_table('astromorphlib/gz_desi_deep_learning_catalog_friendly.parquet')
df = df.sort_values(DESI.MERGING.MERGER.fr, ascending=False, axis='index')
df = df.iloc[:200]
# df = df.iloc[:5]
ls = LegacyService(
  pixscale=0.35, 
  workers=5, 
  replace=True, 
  layer='ls-dr10-grz', 
  width=180, 
  height=180
)
paths = iauname_path(
  ra=df.ra.values, 
  dec=df.dec.values, 
  prefix='stamps/', 
  suffix='.jpg', 
  flat=True
)
ls.batch_cutout(ra=df.ra.values, dec=df.dec.values, save_path=paths)
```
